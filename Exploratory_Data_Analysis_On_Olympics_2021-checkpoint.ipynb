```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
athelete_df = pd.read_csv('athlete_events.csv')
regions_df = pd.read_csv('noc_regions.csv')
```

```python
athelete_df.head()
```

```python
regions_df.head()
```

```python
joined_dataset = pd.merge(athelete_df,regions_df, on='NOC', how='left')
```

```python
joined_dataset.head()
```

```python
joined_dataset.shape
```

```python
for col in joined_dataset.columns:
    if col.startswith('region'):
        joined_dataset.rename(columns = {col:'Region'}, inplace=True)
    
    if col.startswith('notes'):
        joined_dataset.rename(columns = {col:'Notes'}, inplace=True)
```

```python
### check for the top 10 countries with most teams

top_10 = joined_dataset.Team.value_counts().head(10)
```

```python
top_10
```

```python
### plotting on graph the top 10 countries

plt.figure(figsize=(12,6))
plt.title('Top 10 countries')
sb.barplot(x = top_10.index, y=top_10)
```

```python
### age of participants

plt.figure(figsize = (12,6))
plt.title('age of participants')
plt.xlabel('Age')
plt.ylabel('participants')
plt.hist(joined_dataset.Age,bins=np.arange(10,80,2),color='orange')
```
